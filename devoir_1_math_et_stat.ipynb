```python
import matplotlib.pyplot as plt #pour la visualisation
import numpy # va plus vite que math car coder en C
```

```python
def draw_my_function(f, start, end, number_of_points): # f = callback, 
    x_values = numpy.linspace(start, end, number_of_points)
    y_values = f(x_values)
    
    df_dx = numpy.gradient(y_values, x_values)
    
    plt.figure(figsize=(15,6))
    plt.plot(x_values, y_values)
    plt.plot(x_values, df_dx)
    plt.show()
```

```python
start = 0
end = 10
number_of_points = 10000

f = lambda x_values : numpy.exp(-x_values/10) * numpy.sin(x_values) #L'intégrale
g = lambda x_values : numpy.cos(numpy.tanh(x_values)) #La dérivée
```
